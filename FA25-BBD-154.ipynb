```python
N = int(input("Enter a positive integer N: "))
total = 0
i = 1
while i <= N:
    total += i
    i += 1
print("sum from 1 to", N, "is", total)
```

```python
while True:
    try:
        val = float(input("Enter a positive integer number: "))
        if val > 0:
            print(" Thanks, you entered: ", val)
            break
        else:
            print("This is not positive try again.")
    except ValueError:
        print("Invalid input enter a numeric value.")
```

```python
secret = 25
attempt = 0
while True: 
    guess = int(input("guess the secret numbert: "))
    attempt += 1
    if guess == secret:
        print("correct! you took", attempt, "attempt(s). ")
        break
    elif guess > secret:
        print("too high try again.")
    else:
        print("Too low try again.")
```
